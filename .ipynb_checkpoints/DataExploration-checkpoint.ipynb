Then, we can look at all the advertising accounts and the frequency with which they appear.

```python
ad_accounts = Counter(adsDF['author_name'])
ad_accounts
```

Let's also create a percentage breakdown to show how much each advertiser is represented relative to all other advertisers.

```python
total_ads = sum(ad_accounts.values())
print(total_ads)
percentages = {advertiser: (count / total_ads) * 100 for advertiser, count in ad_accounts.items()}
print(percentages)
```

We can see that there are many advertisers that each make up small percentages, so we choose to visualize only the top few for better interpretability.

```python
percentages = {advertiser: (count / total_ads) * 100 for advertiser, count in ad_accounts.items()}
top_percentages = Counter(percentages).most_common(20)

print(top_percentages)
```

```python
# Extracting advertisers and their percentages
advertisers, percentages = zip(*top_percentages)

# Create a visualization for the top 20 advertisers as a percentage of all ads
plt.figure(figsize=(10, 8))
plt.barh(range(len(advertisers)), percentages, align='center', color='skyblue')
plt.yticks(range(len(advertisers)), advertisers)
plt.xlabel('Percentage of Advertisements')
plt.title('Top 20 Advertisers as a Percentage of All Advertisements')
plt.gca().invert_yaxis()  # Invert y-axis to display the most common at the top
plt.savefig('top20ads.png')
plt.show()
```

Now, let's flatten the suggested words column to find the most frequently occuring words among ad and nonad videos.

```python
# Step 1: Split the strings in the 'suggested_words' column by commas
suggested_words_lists = df['suggested_words'].str.split(',')

# Step 2: Flatten the lists
all_words = [word.strip() for sublist in suggested_words_lists for word in sublist]
all_words = [word.lower() for word in all_words]

# Step 3: Count the occurrences of each word
word_counts = Counter(all_words)

# Step 4: Find the top 10 most common words
top_10_words = word_counts.most_common(10)

print(top_10_words)
```

Do the same for ads and non-ads separately to compare them to each other.

```python
# Assuming you have already calculated top_10_words
top_words, word_counts = zip(*top_10_words[1:])

# Create a visualization for the top 10 most frequently occurring suggested words
plt.figure(figsize=(10, 6))
plt.bar(top_words, word_counts, color='skyblue')
plt.xlabel('Suggested Words')
plt.ylabel('Frequency')
plt.title('Top 10 Most Frequently Occurring Suggested Words Among Ads')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top10wordsads.png')
plt.show()
```
